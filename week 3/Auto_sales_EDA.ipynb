## Data Description

Below is the description of each column in the dataset:

| **Column Name**           | **Description**                                                                 |
|---------------------------|---------------------------------------------------------------------------------|
| `ORDERNUMBER`             | Unique identification number assigned to each order.                           |
| `QUANTITYORDERED`         | Number of items ordered in each order.                                         |
| `PRICEEACH`               | Price of each item in the order.                                               |
| `ORDERLINENUMBER`         | Line number of each item within an order.                                      |
| `SALES`                   | Total sales amount for each order (Quantity × Price).                          |
| `ORDERDATE`               | Date on which the order was placed.                                            |
| `DAYS_SINCE_LASTORDER`    | Number of days passed since the customer's last order.                         |
| `STATUS`                  | Status of the order (e.g., Shipped, In Process, Cancelled, etc.).              |
| `PRODUCTLINE`             | Product category to which each item belongs.                                   |
| `MSRP`                    | Manufacturer's Suggested Retail Price for each item.                           |
| `PRODUCTCODE`             | Unique code assigned to each product.                                          |
| `CUSTOMERNAME`            | Name of the customer who placed the order.                                     |
| `PHONE`                   | Customer's contact phone number.                                               |
| `ADDRESSLINE1`            | First line of the customer's address.                                          |
| `CITY`                    | City where the customer is located.                                            |
| `POSTALCODE`              | Postal or ZIP code associated with the customer's address.                     |
| `COUNTRY`                 | Country where the customer is located.                                         |
| `CONTACTLASTNAME`         | Last name of the customer contact person.                                      |
| `CONTACTFIRSTNAME`        | First name of the customer contact person.                                     |
| `DEALSIZE`                | Size of the deal — categorized as "Small", "Medium", or "Large".               |

## Now let's find countries with highest sales

```python
top_countries = df.groupby('COUNTRY')['SALES'].sum().sort_values(ascending=False).head(10)
top_countries
```

### These are the top 10 countries with highest sales , let's visualize it

```python
plt.figure(figsize=(10, 5))
plt.barh(top_countries.index, top_countries.values)
plt.title('Top 10 Countries by Total Sales')
plt.xlabel('Total Sales')
plt.ylabel('Country')
plt.show()
```

## Let's visualize Products by Total Sales

```python
top_products = df.groupby('PRODUCTLINE')['SALES'].sum()
top_products
```

```python
plt.figure(figsize=(10, 6))
plt.pie(top_products.values, labels=top_products.index, autopct='%1.1f%%')
plt.title('Top 10 Products by Total Sales')
plt.axis('equal')# Make the pie chart a circle
plt.legend(loc='upper right')
plt.show()
```

# Interpretation of Results
### Classic Cars dominate:

- With nearly $3.8 million in sales, this product line is the top performer by a large margin.
- It likely has strong customer demand and could be the main revenue driver.

### Vintage Cars are strong too:

- With over $1.8 million, they're the second-highest in sales.
- Indicates sustained interest in collectible or nostalgic products.

### Mid-level performers:

- Motorcycles, Trucks and Buses, and Planes all have sales between $900K–$1.1M.
- These are solid contributors but not leading categories.

### Lower performers:

- Ships and especially Trains have significantly lower sales.
- Could indicate less market interest or limited offerings in these categories.

## Most Frequent Product Lines

```python
# Count occurrences of each product line
product_counts = df['PRODUCTLINE'].value_counts()

# Plot
plt.figure(figsize=(10, 5))
plt.bar(product_counts.index, product_counts.values)
plt.title('Most Frequent Product Lines')
plt.xlabel('Product Line')
plt.ylabel('Number of Orders')
plt.show()
```

## Distribution of Sales

```python
plt.figure(figsize=(10, 5))

plt.hist(df['SALES'], bins=30, color='green', edgecolor='black')
plt.title('Distribution of Sales')
plt.xlabel('Sales Amount')
plt.ylabel('Frequency')
plt.show()
```

# Distribution of MSRP

```python
plt.figure(figsize=(10, 6))
plt.hist(df['MSRP'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of MSRP')
plt.xlabel('MSRP')
plt.ylabel('Count')
plt.show()
```

## Frequency of Deal Sizes (Small, Medium, Large)

```python
# Count deal sizes
deal_counts = df['DEALSIZE'].value_counts()

plt.figure(figsize=(10, 5))
plt.bar(deal_counts.index, deal_counts.values, color='skyblue')
plt.title('Frequency of Deal Sizes')
plt.xlabel('Deal Size')
plt.ylabel('Number of Orders')
plt.show()
```

# Monthly Sales Trends

```python
# Convert ORDERDATE to datetime
df['ORDERDATE'] = pd.to_datetime(df['ORDERDATE'])

# Group by year and month
df['YearMonth'] = df['ORDERDATE'].dt.to_period('M')
monthly_sales = df.groupby('YearMonth')['SALES'].sum()

# Plot
plt.figure(figsize=(12, 5))
monthly_sales.plot(kind='line', marker='o', color='purple')
plt.title('Monthly Sales Trends')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.grid(True)
plt.show()
```

#  How DAYS_SINCE_LASTORDER Affects Sales

```python
plt.figure(figsize=(10, 5))
plt.scatter(df['DAYS_SINCE_LASTORDER'], df['SALES'], alpha=0.5, color='teal')
plt.title('Days Since Last Order vs Sales')
plt.xlabel('Days Since Last Order')
plt.ylabel('Sales')
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Heatmap of Total Sales by Country and Deal Size

```python
import seaborn as sns
# Pivot table: rows = COUNTRY, columns = DEALSIZE, values = sum of SALES
heatmap_data = df.pivot_table(index='COUNTRY', columns='DEALSIZE', values='SALES', aggfunc='sum', fill_value=0)

plt.figure(figsize=(14, 8))
sns.heatmap(heatmap_data, annot=True, fmt='.0f', cmap='YlGnBu', linewidths=0.5, linecolor='gray')

plt.title('Total Sales by Country and Deal Size', fontsize=16)
plt.xlabel('Deal Size')
plt.ylabel('Country')
plt.show()
```

### Heatmap Interpretation
- USA is the top-performing country with the highest sales across all segments.

- France, Spain, and UK also show strong sales, especially in the middle segment.

- Belgium and Switzerland have zero or very low sales in some categories.

- The middle column shows peak sales overall — possibly the most successful product category or time period.

- Focus on high-performing regions and investigate low or missing sales areas.

# Top 10 Customer

# Sales Distribution in Product Line

```python
product_lines = df['PRODUCTLINE'].unique()
data_to_plot = [df[df['PRODUCTLINE'] == line]['SALES'] for line in product_lines]

plt.figure(figsize=(13, 6))
plt.boxplot(data_to_plot, labels=product_lines, patch_artist=True,boxprops=dict(facecolor='lightblue'), medianprops=dict(color='red'))

plt.title('Sales Distribution by Product Line')
plt.xlabel('Product Line')
plt.ylabel('Sales')
plt.show()
```

#  Interpretation:
## Classic Cars:

 - Has a high median sales value.

 - Shows wide variability in sales (tall box + long whiskers).

 - Many outliers — indicating frequent high-ticket sales.

## Trucks and Buses & Vintage Cars:

 - Also show a fairly wide spread in sales.
 
 - Lots of outliers, meaning some orders are significantly above the norm.

## Motorcycles, Planes, Ships, and Trains:

 - Have tighter interquartile ranges (IQR), indicating more consistent sales.

 - Still contain moderate outliers.
